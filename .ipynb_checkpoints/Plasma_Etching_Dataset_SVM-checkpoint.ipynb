## Problem Statement

The task here is to develop empirical models that characterize the behavior of a semiconductor fabrication process, in particular, an ion-assisted plasma etching process, using SVM.

Plasma modeling from a fundamental physical standpoint has had limited success. Current physics-based models attempt to derive self-consisted solutions to first-principle equations involving continuity, momentum balance, and energy balance inside a high frequency, high intensity electric field. This is normally accomplished through expensive numerical simulation methods that are subject to many simplifying assumptions, and are unacceptably slow. Since the complexity of practical plasma processes at the equipment level is presently ahead of theoretical comprehension, other efforts have focused on empirical approaches to plasma modeling.

The task here is to design and train an "optimal" SVM model to recognize the inter-relationships between the process input variables and outputs, using the 51 input-output data pairs provided.

```python
# Setting values of input data to X and output data to y

X = np.asarray(normalized_X)
y = np.asarray(normalized_y)
print("Shape of input data: ", X.shape)
print("Shape of output data: ", y.shape)
```

```python
# Extracting output values

y_1 = y[:, 0]
y_2 = y[:, 1]
y_3 = y[:, 2]
y_4 = y[:, 3]
```

```python
# Hyper-parameter space to optimize SVM

# Parameters options
parameter_space = {'kernel': [ 'rbf', 'polynomial'], 'gamma': [0.001, 0.01, 0.1, 1], 'C': [0.001, 0.01, 0.1, 1, 10]}

# Cross-validation split
cv = KFold(51, shuffle=True, random_state=1000)

# Finding best parameters
clf = GridSearchCV(SVR(),parameter_space, cv = cv)

# Preparing list of outputs values
y_list = [y_1, y_2, y_3, y_4]

# Fit model to find best parameters to each output value column
for columns in y_list:
    clf.fit(X, columns)
    print(clf.best_params_)
```

```python
# Fit model to find best parameters to each output value column

predict_list = []
for columns in y_list:
    clf.fit(X, columns)
    predict = clf.predict(X)
    print(predict)
    predict_list.append(predict)
```
